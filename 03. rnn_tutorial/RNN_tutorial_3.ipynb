## **Tutorial 3 : Predict next even number using Keras**

#### **Keras를 이용하여 다음에 오는 짝수를 구하는 RNN 모델을 만들어보자.**

```python
import numpy as np
import tensorflow as tf

from sklearn.model_selection import train_test_split
from keras.layers import SimpleRNN, Dense
from keras.models import Sequential
from keras.optimizers import Adam
```

#### **Hyper Parameters**

```python
SEQUENCE_NUMBER = 15
HIDDEN_SIZE = 20
BATCH_SIZE = 10
EPOCHS = 2700
DIM_INPUT = 1
DIM_OUTPUT = 1
```

#### **Data creation**

```python
data = [i*2 for i in range(1, 101)]
```

```python
input_seq = []
targets = []
for i in range(len(data) - SEQUENCE_NUMBER):
    input_seq.append(data[i:i+SEQUENCE_NUMBER])
    targets.append(data[i+SEQUENCE_NUMBER])
```

```python
#Resahape : (samples, time steps, features)
input_seq = np.array(input_seq).reshape(len(input_seq), SEQUENCE_NUMBER, DIM_INPUT)
targets = np.array(targets).reshape(len(targets), DIM_OUTPUT)
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_seq, targets, test_size=0.1)
```

#### **Model**

- The `Sequential` model is a linear stack of layers. 

```python
model = Sequential()
```

- Easy to implement

```python
model.add(SimpleRNN(HIDDEN_SIZE, input_shape=(SEQUENCE_NUMBER, DIM_INPUT), kernel_initializer='normal'))
model.add(Dense(DIM_OUTPUT, activation='linear', kernel_initializer='zero'))
adam_optimizer = Adam()
model.compile(adam_optimizer, loss='mean_squared_error')
model.fit(x=X_train, y=y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, shuffle=True, validation_data=(X_test, y_test))
```

```python
model.predict(input_seq)
```
